```python
import requests
import pandas as pd
import pickle
```

```python
params = {'limit':1000}
requests.get('https://api.openaq.org/v1/countries',params).json()
```

```python
params = {'country':'US', 'limit':10000}
requests.get('https://api.openaq.org/v1/cities',params).json()
```

```python
params = {'city':'New York'}
requests.get('https://api.openaq.org/v1/locations', params=params).json()
```

```python
params = {'country': 'US', 'location': 'Bklyn - PS274',  'limit': 10000}
response = requests.get('https://api.openaq.org/v1/measurements', params=params)
```

```python
brooklyn_data = response.json()
```

```python
brooklyn_data
```

```python
measurements = [result['value'] for result in brooklyn_data['results']]
dates = [pd.to_datetime(result['date']['utc']) for result in brooklyn_data['results']]
```

```python
date_measure = {result['date']['utc']: result['value'] for result in brooklyn_data['results']}
date_measure['2019-06-27T21:00:00Z']
```

```python
from collections import OrderedDict
date_measure_sorted = dict(OrderedDict(sorted(date_measure.items(), key = lambda x: x[0])))
```

```python
dates = [pd.to_datetime(date) for date in list(date_measure_sorted.keys())]
measurements = list(date_measure_sorted.values())
```

```python
with open('data/brooklyn.p', 'wb') as write_file:
    pickle.dump([dates, measurements], write_file, )
```
